Niech $\; K \;$ oznacza zbiór liczb zespolonych postaci $\quad x + iy \quad$ gdzie $\quad x, y \in \{-3, -2, -1, 0, 1, 2, 3\}$.


Przez $\; G \;$ natomiast oznaczmy zbiór pierwiastków z jedności stopnia $\; 4 \;$ , czyli $\; G = \{1, i, -1, -i\} \;$.

Jak wiadomo, $\; G \;$ ze  zwykłym mnożeniem jest grupą cykliczną. Jako jej generator możemy wziąć liczbę $\;i \;$.


Zbiór $\; K \;$ reprezentuje nasz kwadrat-ozdobę, a grupa $\; G \;$ reprezentuje obroty przekształcające ten kwadrat w siebie.



Gdy poddajemy $\; K \;$ obrotowi $\; a \in G \;$ to dowolny punkt $\; p \in K \;$ przechodzi na $\; ap \in K \;$.

Łatwo sprawdzić, że $\; G \;$ spełnia warunki definicji pojęcia "działania" w zbiorze $\; K \;$.

Dalej, weźmy pod uwagę zbiór wszystkich dwuelementowych podzbiorów zbioru $\; K \;$ i oznaczmy go przez $\; P \;$:
$$P=\{\; \{p_1, p_2\} : p_1, p_2 \in K , p_1 \neq p_2 \; \} \: .$$

Każdemu $\; \{p_1, p_2\} \in P \;$ odpowiada wzajemnie jednoznacznie ozdoba, w której pola $\; p_1, p_2 \;$ są pomalowane na żółto, a reszta na niebiesko.  

Dla wygody nazwijmy więc elementy zbioru $\; P \;$ "ozdobami".

Przy obrocie $\; a \in G \;$ ozdoba $\; \{p_1, p_2\} \in P \;$ przechodzi na ozdobę $\; \{ap_1, ap_2 \} \in P$.

Widać łatwo, że $\; G \;$ działa w $\; P \;$. Czyli $\; G \;$ rozkłada $\; P \;$ na orbity. 

Z treści zadania z kolei wynika, że dwie ozdoby $\; \{p_1, p_2\}, \: \{p_1', p_2'\}  \;$ są równoważne gdy $\quad \{ ap_1, ap_2 \} = \{ p_1', p_2' \} \quad$ dla pewnego $\; a \in G$,

a więc równoważne ozdoby należą do tej samej orbity. 

Równoważnych ozdób jest więc tyle samo ile jest różnych orbit działania $\; G \;$ w zbiorze $\; P \;$.

Jak wiadomo, liczba elementów orbity jest dzielnikiem rzędu działającej w zbiorze grupy, a ponieważ $\; |G|=4 \;$ więc w naszym przypadku 

orbity mogą być jedynie $\; 1, 2 \;$ lub  $\; 4 \;$ - elementowe.

Zauważmy od razu, że w rozważanym rozkładzie nie ma orbit 1-elementowych tzn. w $\; P \;$ nie ma punktu stałego względem działania grupy $\; G \;$.

Istotnie, jeśli np. $\; \{p_1, p_2\} \;$ byłby punktem stałym to mielibyśmy $\quad \{p_1, p_2\} = \{ap_1, ap_2 \} \quad$ dla każdego $a \in G$,

 $\qquad$ $\qquad$ w szczególności dla $\; a= -1, i \quad$ byłoby : $\{-p_1, -p_2\} = \{ ip_1, ip_2 \}\: .$

Ponieważ $\; p_1 \neq p_2 \;$ więc $\; p_1 \neq 0 \;$ lub $\; p_2 \neq 0 \;$. Niech więc będzie $\; p_1 \neq 0 \;$, wówczas $\; -p_1 \neq ip_1 \;$ 

a więc musi być : $\; -p_1 = ip_2 \quad \text{oraz} \quad -p_2=ip_1\;$
co daje sprzeczność : $\quad p_1=-p_1 \:$.

 Orbity mogą być więc tylko $2$ lub $4$ elementowe.

Oznaczmy przez $\; k_2, k_4 \;$ liczby orbit odpowiednio $2$ i $4$ elementowych. 

Z powyższych rozważań wynika teraz, że szukana przez nas liczba wszystkich różnych ozdób jes równa $\; k_2 + k_4 \;$.

Znajdźmy wartość tej sumy .




Z określenia zbioru $P$ mamy $\; |P|= \binom{|K|}{2} \;$ a z określenia zbioru $K$ mamy $\; |K|=7^2=49$ . Czyli $\; |P|= \binom{49}{2} = 24 \cdot 49 \;$

Ponieważ $P$ został rozbity na rozłączne klasy-orbity $2$ i $4$ elementowe, więc zgodnie z powyższymi oznaczeniami możemy napisać :

$$2\cdot k_2 + 4 \cdot k_4 = 24 \cdot 49 \; ,$$

$$ \text{skąd} \quad k_2 + k_4 = 294 + \frac{k_2}{2} \; .$$

Wystarczy zatem, że znajdziemy wartość $\; k_2 \;$ czyli liczbę wszystkich orbit 2-elementowych względem działania $\; G \;$ w $\; P \;$.

Ponieważ $\; G = <i> \;$ więc orbitą elementu $\; \{p_1, p_2\} \in P \;$ względem działania $\; G \;$ w $P$ jest zbiór postaci: 

$$\{\;\{\;i^np_1 ,\; i^np_2\;\}\; ; n = 0, 1, 2, 3\;\}\;.$$

Wybierzmy pewien element $\;\{p_2, p_2\}  \in P \;$, którego orbita jest 2-elementowa, a więc musi mieć ona postać: 



$$\quad \{\;\{p_1, p_2\}, \{ip_1, ip_2\}\;\}$$


( istotnie, jeśli bowiem byłoby $\;\{p_1, p_2\} = \{ip_1, ip_2\}\;$ to ze względu na $\; G =<i> \;$ powyższa orbita 

byłaby 1-elementowa a wiemy już, że takich nie ma ).


Stąd element $\; \{-p_1, -p_2\} = \{i^2p_1, i^2p_2\}\;$ należy do tej orbity, a więc :

$$\{-p_1,-p_2\}=\{p_1,p_2\} \quad \text{lub} \quad \{-p_1,-p_2\}=\{ip_1, ip_2\}\;.$$

Wcześniej pokazaliśmy już, że druga z tych równości prowadzi do sprzeczości. 

Mamy więc $\; \{p_1, p_2\} = \{-p_1, -p_2\}\;$, a ponieważ $\; p_1 \neq p_2 \;$ więc np. $\; p_1 \neq 0$ czyli $\;p_1 \neq -p_1$

a zatem $\;p_1=-p_2\;$ .

Widzimy teraz, że jeśli orbita jest 2-elementowa to jest ona zbiorem postaci : $\;\{\;\{p, -p\} ,\;\{ip, -ip\}\;\} ,\quad p \in K-\{0\}.$ 

W drugą stronę, jeśli $\;p \in K-\{0\}\;$ to orbita elementu $\;\{p , -p\} \;$ jest 2-elementowym zbiorem powyższej postaci

co wynika łatwo z faktu, że $\;\{\;i^2p\;, -i^2p\;\}=\{\;p ,\; -p\;\}\;$ oraz $\;\{\;i^3p,\;-i^3p\;\}=\{\;ip,\;-ip\}$.

Możemy więc określić następującą surjekcję $\quad orb: K-\{0\} \to \{ orbity \;2\;elementowe\} \quad$ wzorem:

$$orb(p) = \{\;\{\;p,\;-p\},\;\{ip,\; -ip\}\;\}.$$

Wprost z definicji tego odwzorowania wynika, że : $\quad orb^{-1}(orb(p)) = \{\;p,\;-p,\;ip,\;-ip\}\quad$ dla $\;p \in K-\{0\}\;$, a więc 

przeciwobrazy odwzorowania $orb$ są 4-elementowe. Poza tym są one rozłączne i sumują się do zboru $\;K-\{0\}\;$ tzn. 

$$\bigcup_{o}orb^{-1}(o) = K-\{0\}$$

gdzie $\;o\;$ przebiega zbiór wszystkich 2-elementowych orbit działania $\;G\;$ w $\;P\;$.

Liczbę wszystkich takich orbit oznaczyliśmy jako $\;k_2\;$, i pamiętamy, że $\;|K|=49$, stąd i z powyższej równości: $\quad 4k_2=48$ czyli $\;k_2=12$.



Możemy wreszcie obliczyć szukaną wartość : $\quad k_2 + k_4 = 294 + \frac{k_2}{2}=300$.
