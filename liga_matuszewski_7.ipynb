##### Dla $\; n \ge 2 \;$ oznaczmy przez $\; x_n \;$ jedyne dodatnie rozwiązanie równania:
##### $$x + x^2 + ... + x^n =1 \; .$$ 
##### Znajdź $\; \lim_{n \to \infty} x_n \;$.

---

Dla każdego $\; n \ge 2 \;$ określmy funkcję:
$$ 
f_n(x) = x + x^2 + ... + x^n \; . 
$$ 

Łatwo widać, że każda taka funkcja jest ściśle rosnąca na zbiorze $\; x \ge 0 \;$, ze względu na:

$$\forall_{ k \ge 1} \; , \; 0 < a < b \;  \Rightarrow \; a^k < b^k \; .$$

Równie oczywista jest ciągłość każdej z tych funkcji, jako że ciągłe są funkcje postaci $\; x \mapsto x^k \; , k=1, 2, ... \; .$

Dla każdego $\; n \ge 2 \;$ zatem $\; f_n \;$ jest ciągłą bijekcją przedziału $\; [0, \infty) \;$, a więc istnieje

dokładnie jedno $\; x_n \ge 0 \;$ takie, że $\; f_n(x_n)=1 \;$.

Ponadto mamy przy każdym $\; n \ge 2 \;$:
$$ f_n(0)=0 < 1 = f_n(x_n) < n = f_n(1) \; ,$$

a stąd i ze ścisłej monotoniczności wynika: $\; 0 < x_n < 1 \; , n \ge 2$, a więc ciąg $\; (x_n) \;$ jest ograniczony.

Zauważmy dalej, że jest on również malejący.

Istotnie, wybierzmy jakeś $\ n \ge 2 \;.$ 
Ponieważ dla $\; x > 0 \;$ jest oczywiście:
$$
f_n(x)=x + x^2 + ... + x^n < x + x^2 + ... + x^n + x^{n+1} = f_{n+1}(x) \; ,
$$

więc $\quad f_n(x_n) < f_{n+1}(x_n) \;$ , co wraz z równościami $\; f_{n+1}(x_{n+1})=1=f_n(x_n) \;$ daje nam:
$$
f_{n+1}(x_{n+1}) < f_{n+1}(x_n) \;
$$ 

a stąd ponieważ $\; f_{n+1} \;$ jest, jak już wiemy, ściśle rosnąca, mamy $ \; x_{n+1}<x_n \;$. 

Z podstaw analizy wiadomo, że ciąg $\; (x_n) \;$ jako ograniczony i monotoniczny musi być zbieżny.

Oznaczmy przez $\; g \;$ jego granicę.

W celu znalezienia wartości tej granicy, przekształćmy równanie (dla dowolnie ustalonego $\; n \ge 2 \;$):

$$
\begin{aligned} 
f_n(x)&=x+x^2+...+x^n \\ 
(1-x)f_n(x)&=(1-x)(x+x^2+...+x^n) \\
(1-x)f_n(x)&=x - x^{n+1}
\end{aligned} 
$$

skąd po podstawieniu $\; x_n \;$ w miejsce zmiennej $\; x \;$ i uwzględnieniu $\; f_n(x_n)=1 \;$ dostaniemy:

$$
(1-x_n)=x_n - x_n^{n+1}
$$

Ponieważ $\; 0<x_n<1 \;$ dla każdego $\; n \ge 2 \;$ i ciąg $\;  (x_n) \;$ maleje, więc istnieje $\; \lambda \in (0; 1) \;$ takie, że
$$
x_n < \lambda \quad ,n \ge 2
$$

a więc $\quad 0 < x_n^{n+1} < \lambda^{n+1} \;$.

Jest jasne, że $\; \lambda^{n+1} \to 0 \;$. Z tw. o trzech ciągach mamy zatem: $\; x_n^{n+1} \to 0$ 

i możemy  w ostatnim równaniu przejść do granicy wraz z $\; n \to \infty \;$. Otrzymamy:

$$
1 - g = g
$$

$$
g=\frac{1}{2}
$$

### Odpowiedź: $$\lim_{n \to \infty}x_n = \frac{1}{2}$$
