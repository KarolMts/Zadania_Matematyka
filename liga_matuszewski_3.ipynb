# Wykaż, że rozwiązaniem nierówności:$$\frac{1}{x-1}+\frac{2}{x-2}+...+\frac{500}{x-500}\ge\frac{125}{2}$$ jest suma rozłącznych przedziałów o łącznej długości $2004$.

Oznaczmy $\quad f=\sum_{k=1}^{500}\frac{k}{x-k} \;$.

Funkcja ta jest określona i ciągła na zbiorze $\quad \mathbb{R} \setminus \{1, 2, ..., 500\} \quad$ co łatwo widać z jej określenia jako sumy ułamków prostych.

Chcemy opisać i zmierzyć zbiór tych $x \in \mathbb{R}$, dla których $f(x) \ge \frac{125}{2} \;$.

Dla $\quad x < 1 \quad $ jest $\quad \frac{k}{x-k} < 0 \quad$ dla każdego $k=1, 2,..., 500 \quad$ a więc również $\quad f(x)<0 \quad$ i tymbardziej $\quad f(x)<\frac{125}{2} \;$.

Następnie sprawdźmy przebieg funkcji $f$ w przedziale $(k, k+1)$ dla pewnego ustalonego $k=1, 2,... , 499 \;$.

Łatwo zauważyć, że:
$$\lim_{x \to k^{+}}f(x)=\lim_{x \to k^{+}}\frac{k}{x-k}=+\infty \; \text{,} $$ $$\lim_{x \to {(k+1)}^{-}}f(x)=\lim_{x \to {(k+1)^{-}}}\frac{k+1}{x-(k+1)}=-\infty \; \text{,}$$ $$f'(x)=-\sum_{k=1}^{500}\frac{k}{(x-k)^2}<0 \quad dla \quad k<x<k+1 \; \text{.} $$





Widać więc, że nasza funkcja jest ciągła w przedziale $\; (k, k+1) \;$, czyli ma własność Darboux , i maleje od $\; +\infty \;$ do $\; -\infty \;$ w tym przedziale . 

Istnieje zatem takie $\quad c_{k} \in (k, k+1) \quad$ , że:

 $$ f(c_{k})=\frac{125}{2} \; \text{,}$$ $$\quad f(x) \ge \frac{125}{2} \quad \text{dla} \quad k<x \le c_{k} \; \text{,}$$ $$ f(x)<\frac{125}{2} \quad \text{dla} \quad c_{k}<x<k+1 \; \text{.}$$

Pozostało jeszcze sprawdzić przebieg funkcji w przedziale $(500, +\infty) \;$.

Mamy znowu łatwe do sprawdzenia zależności:$$\lim_{x \to {500}^{+}}f(x)=\lim_{x \to {500}^{+}}\frac{500}{x-500}=+\infty \; \text{,}$$
$$\lim_{x \to +\infty}f(x)=0 \; \text{,}$$ $$f'(x)<0 \quad \text{dla} \quad x>500 \; \text{.}$$

Funkcja jest ciągła w przedziale $(500, +\infty)$, więc tu też ma własność Darboux, i maleje od $\; +\infty \;$ do $\; 0 \;$ w tym przedziale.

Istnieje więc takie $\; c_{500} \in (500, +\infty) \;$,że:


$$f(c_{500})=\frac{125}{2} \; \text{,}$$
$$f(x) \ge \frac{125}{2} \quad \text{dla} \quad 500<x \le c_{500} \; \text{,}$$
$$f(x)<\frac{125}{2} \quad \text{dla} \quad x>c_{500} \text{.}$$

W podsumowaniu powyższych rozważań stwierdzamy, że zbiór tych wszystkich $x \in \mathbb{R} \;$, dla których $f(x) \ge \frac{125}{2} \;$ jest postaci:
$$\bigcup_{k=1}^{500}(k, c_k]$$

przy czym przedziały $(k, c_k] \; \text{,} \; k=1, 2, ..., 500 \; \text{,}$ są parami rozłączne.

Łączna długość tych przedziałów jest więc sumą postaci:$$\sum_{k=1}^{500}(c_k - k) \; \text{.}$$ 

Pozostało nam znaleźć wartość liczbową tej sumy.

W tym celu zauważmy, że $\quad f(x) = \frac{w'(x)}{w(x)} \quad$ gdzie:





$$w(x)=(x-1)(x-2)^{2}...(x-500)^{500} \; \text{,}$$
$$w'(x)=\frac{w}{(x-1)}+\frac{w}{(x-2)}+...+\frac{w}{(x-500)} \; \text{.}$$

Stąd jeśli $\quad f(c_{k})=\frac{125}{2} \quad$ to również $\quad \frac{w'(c_{k})}{w(c_{k})} - \frac{125}{2} = 0 \quad$ \(oczywiście $w(c_{k}) \neq 0$\) 

a więc $\quad 2w'(c_{k}) - 125w(c_{k})=0$ .

Czyli wyżej określone punkty $\quad c_{1}, c_{2}, ..., c_{500} \quad$  należą do zbioru pierwiastków wielomianu $\quad 2w - 125w' \quad$. 

Dalej, ponieważ $\; k \;$ jest $\; k$-krotnym pierwiastkiem dla $\; w \;$ więc $\; k \;$ jest $\; (k-1)$-krotnym pierwiastkiem dla $\; w'$ 

a więc liczby $\quad 2, 3, ..., 500 \quad$ są pierwiastkami wielomianu $\quad 2w - 125w' \quad$ o krotnościach $\quad 1, 2, ..., 499 \quad$ odpowiednio.

Z wcześniejszych rozważań wiemy, że liczby $\quad c_{k} \quad (k=1, 2, ...,500)\quad$ są parami różne oraz żadna z nich nie należy do zboru ${2, 3, ..., 499}$.  

Widzimy zatem, że liczby $\quad c_{1}, c_{2}, ..., c_{500}, 2, 3, ..., 500 \quad$ są różnymi pierwiastkami wielomianu $2w'-125w$.

Nie wiemy jednak jeszcze czy wyczerpują one zbiór wszystkich pierwiastków ani też nie znamy krotności pierwiastków $c_{k}$.




Policzmy je wszystkie z uwzględnieniem krotności pierwiastków $2, 3, ..., 500$. 

Mamy $\; 500 \;$ pierwiastków postaci $\; c_{k} \;$ oraz $\quad \sum_{k=2}^{500}
(k-1) \quad$ pozostałych. 

Ich suma $\quad 500+\sum_{k=2}^{500}(k-1) \quad$ nie może być oczywiście większa niż stopień wielomianu $2w'-125w$.

Oznaczmy $\; n=st(2w'-125w)$. Jest jasne, że $\quad n=st(w)=\sum_{k=1}^{500}k$ .

Ale $\quad 500+\sum_{k=2}^{500}(k-1)=\sum_{k=1}{500}=n \quad$ tzn. wyżej obliczona liczba pierwiastków jest równa stopnowi wielomianu $2w'-125w$.

Czyli liczby $\quad c_{1}, c_{2}, ..., c_{500}, 2, 3, ..., 500 \quad$ wyczerpują zbiór wszystkich pierwiastków naszego wielomianu a pierwiastki $\; c_{k} \;$ są jednokrotne.

Powyższy wynik daje nam podstawy do skorzystania ze wzorów Viet'a w celu obliczenia sumy $\quad \sum_{k=1}^{500}(c_{k}-k)$ .

Oznaczmy $\quad w=a_{n}x^{n}+a_{n-1}x^{n-1}+...$

wówczas $\quad w'=na_{n}x_{n-1}+...$

oraz $\quad 2w'-125w=-125a_{n}x^{n}+(2na_{n}-125a_{n-1})x^{n-1}+...$

Zsumujmy wszystkie pierwiastki wielomianu $\; 2w-125w' \;$. Zgodnie z wcześniejszymi wnioskami jest to liczba: $\quad \sum_{k=1}^{500}c_{k}+\sum_{k=2}^{500}k(k-1)$ .



Stąd i ze wzorów Viet'a dostajemy równanie: $$-\frac{2na_{n}-125a_{n-1}}{-125a_{n}}=\sum_{k=1}^{500}c_{k}+\sum_{k=2}^{500}k(k-1)$$

Teraz wystarczy podstawić odpowiednie wartości pod $\; n , \; a_{n-1} ,\; a_{n} $ .

Już wiemy, że: $\quad n=st(w)=\sum_{k=1}^{500}k$ .

Z kolei z określenia $ \; w \;$ mamy łatwo: $\quad a_{n}=1$ .

Gdy natomiast zastosujemy wzory Viet'a do $\; w \;$ to otrzymamy: $\quad a_{n-1}=-\sum_{k=1}^{500}k^2$ .

Podstawiamy i dostajemy:$$\sum_{k=1}^{500}c_{k}+\sum_{k=2}^{500}k(k-1)=-\frac{2\sum_{k=1}^{500}k-125(-\sum_{k=1}^{500}k^2)}{-125}$$

$$\sum_{k=1}^{500}c_{k}+\sum_{k=2}^{500}k^2+500-\sum_{k=2}^{500}k-500=\frac{2}{125}\sum_{k=1}^{500}k+\sum_{k=1}^{500}k^2$$

$$\sum_{k=1}^{500}c_{k}+\sum_{k=1}^{500}k^2-\sum_{k=1}^{500}k=\frac{2}{125}*\frac{500*501}{2}+\sum_{k=1}^{500}k^2$$

Gdzie po prawej skorzystaliśmy ze znanego wzoru na skończoną sumę kolejnych wyrazów ciągu arytmetycznego.

Po łatwym przekształceniu otrzymujemy wreszcie szukany wynik:$$\sum_{k=1}^{500}(c_{k}-k)=2004$$
