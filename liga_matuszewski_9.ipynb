#### Gracz rzuca sześcienną kością (z liczbami oczek 1, 2, ..., 6) tak długo, aż łączna

#### suma oczek nie będzie równa przynajmniej $\; n \;$. Oznaczmy przez $\; p(n) \;$ 

#### prawdopodobieństwo, że będzie ona wynosiła dokładnie $\; n \;$. Wyznacz:

#### $$\lim_{n \to \infty} p(n) \; .$$

#### Możesz założyć, że ta granica istnieje.

---


Oznaczmy przez $\; X_i \; (i=1, 2, ...) \;$ wynik i-tego rzutu, a przez $\; S_k \; (k=1, 2, ...) \;$ łączną sumę oczek otrzymaną w $\; k \;$ rzutach.

Czyli: $\quad S_k = X_1 + ... + X_k \;$.

Równość $\; S_k=n \;$ oznacza zdarzenie:  *"w $\; k \;$ rzutach łączna suma oczek wyniosła $\; n \;$"* . Niech $\; p(S_k=n) \;$ oznacza 

prawdopodobieństwo jego zajścia.

Przy ustalonym $\; n \;$ ale dla różnych $\; k \;$ zdarzenia te są rozłączne a suma ich wszystkich składa się na zdarzenie : 

*"suma oczek wyniosła dokładne $\; n \;$"* , stąd :

$$p(n)=\sum_{k=1}^{\infty} p(S_k=n) \qquad (1)$$

przy czym jest jasne, że prawie wszystkie składniki tej sumy są zerami ponieważ liczba rzutów nie może być większa niż suma oczek 

tzn. $\; p(S_k=n)=0 \;$ gdy $\; k>n \;$.

Dalej zauważmy, że zdarzenie $\; S_k=n \;$ przy $\; k>1 \;$ jest sumą rozłącznych zdarzeń polegających na tym, że: 

*"w $\; k-1 \;$ rzutach wypadło łącznie w sumie $\; n-i \;$ oczek oraz w $k$-tym rzucie wypadło $\; i \;$ oczek"* 

gdzie $\; i=1, 2, 3, 4, 5, 6 \;$, czyli:

$$\quad p(S_k=n)=\sum_{i=1}^{6}p(S_{k-1}=n-i \; ; \; X_k=i) \; ,$$

a ponieważ wynik $\; k$-tego rzutu nie zależy od poprzednich $\; k-1 \;$ wyników tzn. $\; p(X_k=i \; | \;  S_{k-1}=n-i)=p(X_k=i) \;$ więc, 

zgodnie ze wzorem na prawdopodobieństwo warunkowe możemy napisać :

$$\quad p(S_k=n)=\sum_{i=1}^{6}p(S_{k-1}=n-i) \cdot p(X_k=i)  \; , \quad k=2, 3, ... \; .$$

Podstawiając to do $(1)$ i uwzględniając, że  $\; p(X_k=i)=\frac{1}{6} \;$ dla $\; i=1, 2, 3, 4, 5, 6 \;$,

dostaniemy następujący wzór na $\; p(n) \;$:

$$p(n)=p(S_1=n) + \frac{1}{6} \sum_{k=2}^{\infty}\sum_{i=1}^{6}p(S_{k-1}=n-i)$$
skąd po zmianie porządku sumowania i skorzystaniu ze wzoru $(1)$ otrzymujemy:

$$p(n)= p(S_1=n) + \frac{1}{6}\sum_{i=1}^{6}p(n-i) \; .$$
Jeśli $\; n > 7 \;$ to oczywiście $\; p(S_1=n)=0 \;$ a więc:

$$ p(n)=\frac{1}{6}p(n-1)+ ... + \frac{1}{6}p(n-6) \quad \text{dla} \; n=7, 8, ... \; .$$
Niech z kolei $\; n \in \{1, ..., 6\} \;$ wówczas $\; p(S_1=n)=\frac{1}{6} \;$ , natomiast dla $\; n-i \leq 0 \;$ mamy $\; p(n-i)=0 \;$ co 

wynika ze wzoru $(1)$ ponieważ $\; p(S_k <1)=0\;.$ Mamy zatem:

$$p(n)= \frac{1}{6} + \frac{1}{6}p(n-1) + ... + \frac{1}{6}p(1) \quad \text{dla} \; n=1, ...,6 .$$
Jeśli przyjmiemy, że $\; p(0)=1 \;$ oraz $\; p(i)=0 \;$ dla $\; i<0 \;$ to nie wpłynie to na wzór otrzymany dla $\; n>7 \;$ 

natomiast dla $\; n=1, ..., 6 \;$ możemy napisać:

$$p(n)=\frac{1}{6}p(0) + \frac{1}{6}p(n-1) + ... + \frac{1}{6}p(1)=\frac{1}{6}p(n-1)+ ... + \frac{1}{6}p(0)+ ... + \frac{1}{6}p(n-6) \; .$$
W ten oto sposób otrzymaliśmy rekurencyjny wzór dla ciągu $\; p(n) \;$:

$$ p(n)=0  \quad \text{dla} \; n<0 \; ,$$

$$p(0)=1 \; ,$$

$$ p(n)=\frac{1}{6}p(n-1)+ ... + \frac{1}{6}p(n-6) \quad \text{dla} \; n=1, 2, ... \; .$$

W celu znalezienia granicy tego ciągu posłużymy się aparatem funkcji tworzących (zob. [R] 3.2.12 , 3.5.10). Oznaczmy jako $\; f \;$ funkcję tworzącą naszego ciągu, czyli 

$\; f \;$ oznacza następujący formalny szereg potęgowy:

$$f=\sum_{n=0}^{\infty}p(n)x^n  \qquad (2) \; .$$
Mnożąc powyższy szereg przez wielomian $\; w = 1 - \frac{1}{6}x - ... - \frac{1}{6}x^6 \quad$ (który też jest potęgowym szeregiem formalnym):

$$w \cdot f = (1 - \frac{1}{6}x - ... - \frac{1}{6}x^6) \cdot \sum_{n=0}^{\infty}p(n)x^n $$
i korzystając z określenia mnożenia formalnych szeregów potęgowych :

$$\left( \sum a_nx^n \right) \cdot \left( \sum b_nx^n \right)= \sum \left( \sum_{i+j=n}a_ib_j \right)x^n$$
otrzymujemy równość :

$$w \cdot f=\sum_{n=0}^{\infty} \left(p(n) - \frac{1}{6}p(n-1)- ... - \frac{1}{6}p(n-6) \right)x^n$$
w której zgodnie z powyższym wzorem rekurencyjnym na wyrazy ciągu $\; p(n) \;$ jest:

$$p(n) - \frac{1}{6}p(n-1)- ... - \frac{1}{6}p(n-6)=0 \quad \text{dla} \; n=1, 2, ...$$

$$\text{oraz} \quad p(0)=1 \; .$$


$\text{Dostajemy zatem :} \quad w \cdot f=1$ a więc:

$$f=\frac{1}{w}=\frac{1}{1-\frac{1}{6}x-...-\frac{1}{6}x^6}=\frac{-6}{x^6+...+x -6} \qquad (3) \; .$$
Dalej zauważmy, że wielomian $\; x^6 + ... + x - 6 \;$ ma pierwiastki wszystkie jednokrotne o modułach nie mniejszych od jedności.

Istotnie, oznaczmy ten wielomian przez $\; g=x^6+...+x-6 \;$.Widać od razu, że $\; g(1)=0 \;$ oraz $g'(1)=6+..+1 \neq 0 \;$ 

a więc $\; 1 \; $ jest pierwiastkiem jednokrotnym. 

Niech więc $\; \epsilon \; \neq 1 \;$ będzie innym pierwiastkiem dla $\; g \;$.

Pomnóżmy $\; g \;$ przez jednomian $\; x-1 \;$. Po wymnożeniu otrzymamy:

$$(x-1)g=x^7-7x+1 \; .$$

Jeśli $\; \epsilon \;$ byłby pierwiastkiem wielokrotnym to byłby nim również dla pochodnej $\; \left[(x-1)g\right]'=7x^6-7$ 

a więc mielibyśmy: $ \quad 7\epsilon^6-7=0 \quad \text{czyli} \quad \epsilon^6=1 \; .$

Z drugiej strony mamy:

$$g=x^6+...+x+1-7=\frac{x^7-1}{x-1}-7$$
a stąd, ze względu na $\; \epsilon \neq 1 \; \text{oraz} \; \epsilon^7=\epsilon \;$: 

$$\; g(\epsilon)=\frac{\epsilon^6-1}{\epsilon-1}-7=\frac{\epsilon-1}{\epsilon-1}-7=-6$$
co jest sprzeczne z $\; g(\epsilon)=0 \; .$ Pierwiastek $\; \epsilon \;$ musi być więc jednokrotny.
Dalej, jeśli byłoby $\; |\epsilon| < 1 \;$ to ze względu na $\; g(\epsilon)=0 \;$ dostalibyśmy:

$$6=|\epsilon^6+...+\epsilon| \le |\epsilon^6|+...+|\epsilon|<6$$
czyli sprzeczność. Musi być więc $\; |\epsilon| \ge 1 \; $ a nawet $\; |\epsilon| > 1 \;$ gdyż $\; \epsilon \neq 1 \; .$
Dla wielomianu $\; g \;$ stopnia $\; 6 \;$, zgodnie z wnioskiem z zasadniczego twierdzenia algebry ([M&S] rozdz.VII, par.1.3) istnieje rozkład:

$$g=(x-\epsilon_0)(x-\epsilon_1) ... (x-\epsilon_5)  $$
i według powyższych rozważań możemy powiedzieć, że:

$$\epsilon_0=1 \; ,$$

$$\epsilon_i \neq \epsilon_j \quad \text{gdy} \quad i \neq j \; ,$$

$$|\epsilon_i|>1 \quad \text{dla} \quad i=1, ..., 5 \; .$$
Z elementów algebry wiemy, że funkcję wymierną $\; \frac{1}{g} \;$ możemy przedstawić jako sumę ułamków prostych postaci:

$$\frac{1}{g}= \frac{a_0}{x-\epsilon_0} + \frac{a_1}{x-\epsilon_1}+...+\frac{a_5}{x-\epsilon_5} \qquad (4)$$
dla pewnych stałych $\; a_0, a_1, ..., a_5 \;$ ([M&S] rozdz.VI par.5.2 , [R] 3.5.9). 
Wracając do funkcji tworzącej $\; f \;$ ze wzorów $(2) \; , \; (3) \;$ dostajemy:

$$\sum_{n=0}^{\infty}p(n)x^n=-6 \cdot \left(\frac{a_0}{x-\epsilon_0} + \frac{a_1}{x-\epsilon_1}+...+\frac{a_5}{x-\epsilon_5} \right)$$
gdzie każdy z ułamków prostych możemy przedstawić jako formalny szereg potęgowy tzn.:

$$\frac{a}{x-\epsilon}=-\frac{a}{\epsilon} \cdot \left( \frac{1}{1- \frac{x}{\epsilon}} \right)=-\frac{a}{\epsilon}\sum_{n=0}^{\infty}\left(\frac{1}{\epsilon}\right)^nx^n$$
a stąd po porównaniu współczynników przy $\; x^n \;$ otrzymujemy wzór wyraźny na $\;n$-ty wyraz ciągu $\; p(n) \;$:

$$p(n)=6 \cdot \left[ \frac{a_0}{\epsilon_0} \left(\frac{1}{\epsilon_0}\right)^n+\frac{a_1}{\epsilon_1}\left(\frac{1}{\epsilon_1}\right)^n+...+\frac{a_5}{\epsilon_5}\left(\frac{1}{\epsilon_5}\right)^n\right] \; .$$
Pamiętamy, że $\; \epsilon_0=1 \;$ oraz $\; |\epsilon_i|>1 \;$ dla $\; i=1, ..., 5 \;$ a więc:

$$\lim_{n \to \infty}\left(\frac{1}{\epsilon_i}\right)^n=0 \qquad \text{dla} \; i=1, ..., 5$$
### $$\text{stąd:} \qquad \lim_{n \to \infty}p(n) = 6a_0$$
i pozostało nam więc tylko znaleźć wartość $\; a_0 \; .$
W tym celu przemnóżmy $\; (4) \;$ przez $\; g \;$ i wstawmy $\; \epsilon_0=1 \;$. Otrzymamy:

$$1=a_0\frac{g}{x-1}+a_1\frac{g}{x-\epsilon_1}+...+a_5\frac{g}{x-\epsilon_5} \; \qquad (5).$$
Następnie zauważmy, że :

$$g=x^6+...+x-6=(x-1)(6+5x+4x^2+3x^3+2x^4+x^5)$$
$$\text{a więc} \qquad \frac{g}{x-1}=6+5x+4x^2+3x^3+2x^4+x^5 \; .$$
Stąd, ponieważ $\; \frac{g(1)}{1-\epsilon_i}=0 \;$ dla $\; i=1, ..., 5 \;$, po wstawieniu $\; x=1 \;$ do $\; (5) \;$ otrzymamy:

$$1=a_0\cdot(6+5+4+3+2+1)$$
$$\text{czyli} \qquad a_0=\frac{1}{21}$$
i ostatecznie dostajemy rozwiązanie naszego zadania:

### $$\lim_{n \to \infty}p(n)=\frac{2}{7}$$
