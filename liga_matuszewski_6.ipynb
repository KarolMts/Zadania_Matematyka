#### Dla każdego $\; n \in \mathbb{N}$ oblicz wyznacznik macierzy $$ A=[min\{i, j\}]_{1 \le i, j \le n}$$



---

Dla $n=1$ mamy trywialne $\; det(A)=1$. Niech więc nasza macierz $\; A \;$ ma stopień $\; n > 1 \;$ i oznaczmy przez $\;B = [b_{i, j}]_{1 \le i, j \le n} \;$ macierz 

jaka powstaje z $\; A \;$ w ten sposób, że :

od $\; n$-tej kolumny odejmujemy $\; (n-1)$-szą, następnie 

od $\; (n-1)$-szej odejmujemy $\; (n-2)$-gą 

itd. po kolei od $\; j$-tej kolumny odejmujemy $\; (j-1)$-szą 

aż na końcu od drugiej odejmiemy pierwszą.

Jak wiadomo z elementarnego kursu algebry ([1] tw.4.31), takie przekształcenia nie zmieniają wyznacznika a więc 

$$
det(A)=det(B).
$$

Dalej, z określenia $\; A=[min\{i, j\}] \;$ widzimy, że wyrazy macierzy $\; B \;$ są postaci:

$$
b_{i, j} = min\{i, j\} - min\{i, j-1\} =
\begin{cases}
0 & \text{dla } i > j \\
1 & \text{dla } i \le j
\end{cases}
$$

czyli wszystkie wyrazy powyżej głównej przekątnej są zerami a więc mamy do czynienia z macierzą trójkątną i wyznacznik takiej macierzy 

jest, jak wiadomo ([2] rozdz.IV par.2 tw.2), iloczynem wyrazów głównej przekątnej, a więc zgodnie z powyższym wzorem:

$$
det(B) = b_{1, 1} \; b_{2, 2} \;  ...  \; b_{n, n} = 1
$$

a więc ostatecznie 
$$
det(A)=1.
$$.
