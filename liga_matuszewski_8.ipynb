### Dana jest ściśle rosnąca i surjektywna funkcja $\; g:[0, 1] \to [0, 1] \;$. 

### Wykaż, że dla dowolnego $\; n \in \mathbb{N} \;$:

### $$\sum_{k=1}^{n-1}\left(g\left(\frac{k}{n}\right) +g^{-1}\left(\frac{k}{n}\right)\right) < n- \frac{1}{n} \; ,$$

### gdzie $g^{-1} \;$ oznacza funkcję odwrotną do $\; g \;$.

---

Jest jasne, że $\; g^{-1} \;$ jest także ściśle rosnącą surjekcją.

Funkcje $\; g, g^{-1} \;$ jako monotoniczne, mogą mieć nieciągłości tylko pierwszego rodzaju, a jako surjektywne nie mogą mieć skoków.

Funkcje te są zatem ciągłe a więc i całkowalne na przedziale $\; [0,1]$, (zob. [F1] par.71 pkt.2 oraz [F2] par.298 pkt.1).

Dla dowolnego $\; t \in (0,1] \;$ rozważmy w prostokątnym układzie współrzędnych $\;OXY \;$ 

prostokąt o wierzchołkach w punktach:$\;(0,0),\;(0,t),\;(t,g(t)),\;(0,g(t))\;$,

oraz jego krzywoliniową przekątną : $\;\{\;(x,g(x))\;:\;x\in[0,t]\;\}\;$, która leży całkowicie we wnętrzu naszego prostokąta ( gdyż $\; g \;$ jest bijekcją )

i dzieli go na dwie figury o rozłącznych wnętrzach.

Pole figury leżącej pod tą przekątną wyraża się całką: $\; \int_{0}^{t}g \;$,

zaś pole figury leżącej nad przekątną równe jest całce: $\; \int_{0}^{g(t)}g^{-1} \quad $(zob. [F2] par.338).

Suma tych całek daje nam więc pole rozważanego prostokąta tj. zachodzi wzór:

$$ t \cdot g(t)=\int_{0}^{t}g + \int_{0}^{g(t)}g^{-1} \; \qquad \qquad(*).$$

W szczególności przy $\; t=1 \;$ mamy oczywiście $\; g(1) = 1 \; $ oraz :

$$1=\int_{0}^{1}g + \int_{0}^{1}g^{-1} \qquad \qquad(**).$$

Ponadto zauważmy, że dla dowolnego $\; t \in [0,1] \;$ zachodzi nierówność:
$$\int_{0}^{t}g + \int_{0}^{t}g^{-1} \ge t^{2} \qquad \qquad(***).$$

Istotnie, jeśli $\; t \ge g(t) \;$ to wobec $(*)$ i wobec monotonoczności $\; g^{-1} \;$:

$$\int_{0}^{t}g + \int_{0}^{t}g^{-1} = \int_{0}^{t}g + \int_{0}^{g(t)}g^{-1} + \int_{g(t)}^{t}g^{-1} \ge t \cdot g(t) + [t-g(t)] \cdot g^{-1}(g(t)) = t^{2} \; .$$

Z kolei gdy $\; t \le g(t) \;$ czyli $\; t \ge g^{-1}(t) \;$, to zamieniając w $(*)$ $\; t \;$ na $\; g^{-1}(t) \;$ i biorąc pod uwagę monotoniczność $\; g \;$ mamy:

$$\int_{0}^{t}g+\int_{0}^{t}g^{-1}=\int_{0}^{g^{-1}(t)}g+\int_{g^{-1}(t)}^{t}g+\int_{0}^{t}g^{-1} \ge g^{-1}(t) \cdot t + [t-g^{-1}(t)] \cdot g(g^{-1}(t))=t^{2} \; .$$

Dalej, weźmy dowolne $\; n \in \mathbb{N} \;$ oraz $\; k \in \{1,...,n-1\} \;$.

Ponieważ funkcje $\; g \;, g^{-1} \;$ są ściśle rosnące, dla każdego $\; x \in (\frac{k}{n},\frac{k+1}{n}] \;$ mamy ostrą nierówność:

$$g\left(\frac{k}{n}\right)+g^{-1}\left(\frac{k}{n}\right)<g(x)+g^{-1}(x) \; ,$$



skąd po zastosowaniu operacji $\; \int_{\frac{k}{n}}^{\frac{k+1}{n}} \;$ do lewej i prawej strony otrzymujemy :

$$\frac{1}{n}\cdot \left(g\left(\frac{k}{n}\right)+g^{-1}\left(\frac{k}{n}\right)\right)<\int_{\frac{k}{n}}^{\frac{k+1}{n}}\left(g+g^{-1}\right) \; , $$

a następnie po zastosowaniu operacji $\; \sum_{k=1}^{n-1} \;$ do obu stron dostajemy:

$$$$

$$\frac{1}{n}\sum_{k=1}^{n-1}\left(g\left(\frac{k}{n}\right)+g^{-1}\left(\frac{k}{n}\right)\right)<\int_{\frac{1}{n}}^{1}\left(g+g^{-1}\right)=\int_{0}^{1}\left(g+g^{-1}\right)-\int_{0}^{\frac{1}{n}}\left(g+g^{-1}\right) \; .$$

Korzystając ze związków $(**)$ oraz $(***)$ możemy więc napisać:

$$\frac{1}{n}\sum_{k=1}^{n-1}\left(g\left(\frac{k}{n}\right)+g^{-1}\left(\frac{k}{n}\right)\right)< 1-\frac{1}{n^{2}} \; ,$$

a stąd po pomnożeniu obu stron przez $\; n \;$:

#### $$\sum_{k=1}^{n-1}\left(g\left(\frac{k}{n}\right)+g^{-1}\left(\frac{k}{n}\right)\right)<n-\frac{1}{n} \quad \text{c.b.d.o}$$
