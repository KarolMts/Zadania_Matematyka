# Znajdź wszystkie grupy skończone, które mają dokładnie 3 podgrupy.

Grupa mająca dokładnie 3 podgrupy to grupa, która ma dokładnie jedną podgrupę właściwą.

Niech $G$ będzie taką grupą i oznaczmy przez $H$ jej jedyną podgrupę właściwą oraz przez $e$ jej element neutralny.

**Jeśli $G$ nie jest cykliczna** to dla każdego $a \in G-\{e\}$ podgrupa $\langle a \rangle$ (generowana przez $a$) jest właściwa 
a więc $H=\langle a \rangle$ wobec założenia o $H$.

Mamy więc $\quad H=\bigcup_{a \in G}\langle a \rangle=G \quad$ - sprzeczność.

Grupa $G$ musi być zatem cykliczna, czyli $G=\langle a \rangle$ dla jakiegoś $a \in G$.

Oznaczmy jej rząd przez $n$. Liczba $n$ nie może być pierwsza gdyż w przeciwnym razie nasza grupa byłaby prosta tzn. nie miałaby podgrupy właściwej.

Mamy zatem $n=p*m$ dla pewnej liczby pierwszej $p$ oraz pewnej całkowitej $m > 1$.

Ponadto elementy $a^{p}, a^{m}$ są rzędu $m, p$ odpowiednio czyli podgrupy $\langle a_{p} \rangle, \langle a_{m} \rangle$ mają rzędy $m, p$

gdzie $1<m, p<n$ czyli podgrupy te są właściwe.
Musi być więc $H=\langle a^{p} \rangle=\langle a^{m} \rangle$ czyli $p=m$ skąd $n=p^2$.

Pokazaliśmy, że **jeśli grupa ma dokładnie 3 podgrupy to jest ona cykliczna rzędu $p^2$ dla pewnej liczby pierwszej $p$**.




Pozostało nam pokazać jeszcze, że każda grupa cykliczna rzędu $p^2$ ($p$ - liczba pierwsza) ma dokładnie 3 podgrupy tzn. ma

dokładnie jedną podgrupę właściwą.

Niech $G$ będzie cykliczna rzędu $p^2$ dla pewnej liczby pierwszej $p$, czyli  dla pewnego $a \in G$ mamy 
$G-\{e\}=\{a^k : k=1, 2, ... , p^2-1\}$, i $H$ oznacza jej właściwą podgrupę.

Wiadomo, że *rząd*$H$ dzieli *rząd*$G$ (tw. Lagrange'a) stąd, ponieważ $H$ jest właściwa, mamy *rząd*$H=p$.

Zauważmy, że $H \subset \langle a^p \rangle$. Istotnie, jeśli $a^k \in H$ dla pewnego $k=1, 2,...,p^2-1$ to $p \mid k$, w przeciwnym razie

$k$ i $p$ byłyby względnie pierwsze a to oznacza, że $G=\langle a^k \rangle \subset H $ wbrew temu, że $H$ jest właściwa.

Dalej, jasne jest, że *rząd*$\langle a^p \rangle=p$, stąd, ponieważ *rząd*$H=p$ i $H \subset \langle a^p \rangle$ dostajemy

$$H=\langle a^p \rangle$$

czyli każda podgrupa właściwa grupy $G$ jest równa podgrupie $\langle a^p \rangle$ tzn. $G$ nie ma innych podgrup właściwych poza $\langle a^p \rangle$.

# Grupa ma dokładnie 3 podgrupy wtedy i tylko wtedy gdy jest cykliczna rzędu $p^2$ dla jakiejś liczby pierwszej $p$
