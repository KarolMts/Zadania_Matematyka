# Ala i Bartek bawią się w dokładanie kamyków do pudełka.
# Na początku w pudełku są 2 kamyki. Gracze wykonują ruchy na przemian.

# Jeśli przed ruchem w pudełku jest $x$ kamyków, to gracz musi dołożyć

# dodatnią liczbę kamyków, ale nie więcej niż $x$.

# Wygrywa ten, kto jako pierwszy doprowadzi liczbę kamyków w pudełku do $2026$.

# Zakładając, że grę zaczyna Ala, który z graczy ma strategię wygrywającą?

---

Zdefiniujmy rekurencyjnie następujący ciąg :

$\begin{align} x_0&= 2026  \\ x_{n+1} &= \left\lceil \frac{x_n}{2}-1 \right\rceil . \end{align} $

Z własności funkcji $\; a \mapsto \lceil a \rceil \;$ mamy : $\quad \frac{x_n}{2}-1 \le x_{n+1} < \frac{x_n}{2} \quad ,$ a więc jeśli $\quad x_n > 0 \quad$ to $\quad 0 \le x_{n+1} < x_n\; $. 
Ponadto $\; x_0 > 0 \;$.

Wynika stąd, że istnieje liczba $\; m=max\{n: x_n > 0\} \;$ oraz, że skończony ciąg: $\quad x_0, x_1, ..., x_m \quad$ jest malejący.

Dla $\; n = 0, 1, ..., m \;$ oznaczmy przez $\; \mathbf{F_n} \;$ następujące zdanie:

" Wygrać może gracz, który jako pierwszy doprowadzi liczbę kamyków w pudełku do $\; x_n \;$". ("może" o ile stosuje strategię )

Zdanie $\; \mathbf{F_0} \;$ jest oczywiście prawdziwe jako część zasad gry.

Załóżmy, że prawdziwe jest zdanie $\; \mathbf{F_n} \;$ dla pewnego $\; n \;$, $\; 0 \le n < m \;$, i że jeden z graczy (nazwijmy go graczem $\; \mathbf{W}\;$) doprowadził

liczbę kamyków w pudełku do $\; x_{n+1} \;$. 

Kolejny ruch należy do jego przeciwnika, który dokłada do pudełka pewną liczbę $\; k \;$ kamyków i według 

zasad gry musi być: $\; 1 \le k \le x_{n+1}$. 

Po tym ruchu w pudełku jest $\; x= x_{n+1} + k\quad$ kamyków. 

Ruch należy do $\; \mathbf{W} \;$ i zauważmy, że ma on prawo dołożyć $\quad x_n - x \quad$ kamyków tzn. przestrzega reguły gry
 : $\quad 1 \le x_n - x \le x \;$.

Istotnie, ze wspomnianych już wcześniej zależności : $\; \frac{x_n}{2} -1 \le x_{n+1} < \frac{x_n}{2} \quad$ oraz $\quad 1 \le k \le x_{n+1} \;$ mamy kolejno : 

$\begin{align} x_n - 2 \le 2x_{n+1} \quad &\text{oraz} \quad 2x_{n+1} < x_n \\  x_n \le 2x_{n+1} +2 \le 2x_n + 2k\quad &\text{oraz} \quad x_{n+1} +k \le 2x_{n+1} < x_n \\ x_n \le 2(x_{n+1}+k) \quad &\text{oraz} \quad x_{n+1} +k +1 \le x_n \\ x_n \le 2x \quad &\text{oraz} \quad x+1 \le x_n \end{align}$

$$\text{czyli} \quad 1 \le x_n-x \le x \; .$$

Jeśli teraz gracz $\; \mathbf{W} \;$ dołoży $\quad x_{n}-x \quad$ kamyków to doprowadzi ich liczbę w pudełku do $\; x_n \;$ a więc może wygrać wg. przyjętej prawdziwości zdania $\; \mathbf{F_n} \;$.  

Liczba $\; k \;$ została wybrana dowolnie z dostępnych możliwości : $\; 1, 2, ..., x_{n+1} \;$ a zatem gracz $\; \mathbf{W} \;$ ma wygrywającą 

odpowiedź na każde (legalne) posunięcie oponenta jeśli w poprzednim swym ruchu doprowadził liczbę kamyków w pudełku do $\; x_{n+1} \;$ i jeśli zdanie $\; \mathbf{F_n} \;$ jest prawdziwe.

Stąd wynika prawdziwość zdania $\; {F_{n+1}}\;$.

Na mocy prawa indukcji stwierdzamy, że zdanie $\; \mathbf{F_n} \;$ jest prawdziwe dla każdego $\; n=0, 1, ..., m$.




Aby dowiedzieć się kto ma strategię wygrywającą obliczmy pierwszych $\; m+1 \;$ wyrazów powyższego ciągu (tzn. wszystkie jego dodatnie wyrazy).

Łatwy rachunek daje nam następujący ciąg : $\; 2026,\; 1012,\; 505,\; 252,\; 125,\; 62,\; 30,\; 14,\; 6,\; 2 \; $. Poza tym $\; m=9 \;$.

Skoro w pudełku znajduje się na starcie $\; x_m=2 \;$ kamyków i Ala zaczyna grę, więc Bartek może obrać strategię następującą :

##### jeśli w pudełku znajduje się $\; x_{n+1} \;$ kamyków ($\; 0 \le n < m= 9 \;$) a Ala dokłada do tego $\; k \;$ kamyków , to Bartek w odpowiedzi

##### dokłada $\; x_n - x \;$ kamyków ($\; x = x_n +k \;$) doprowadzając ich liczbę w pudełku do $\; x_n \;$.

Wcześniejsze rozmowanie indukcyjne pokazuje, że postępując w ten sposób Bartek jako pierwszy doprowadzi liczbę kamyków w pudełku do $\; x_0 = 2026 \; $ , a więc wygra grę czyli

# Bartek ma strategię wygrywającą.
